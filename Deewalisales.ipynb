# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', hue='Gender', data=sales_gen)
```

```python
plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouping and summing the data
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Creating the bar plot with a specified palette
sns.barplot(x='Age Group', y='Amount', data=sales_age, palette='Set2')

# Adding labels and title
plt.title('Total Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.tight_layout()       # Adjust layout to prevent clipping
plt.show()
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouping and summing the data
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Creating the bar plot with a specified palette
sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(data=sales_state, x='State', y='Orders', palette='Set2')

# Adding labels and title
plt.title('Total Orders by Top 10 States')
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.tight_layout()       # Adjust layout to prevent clipping
plt.show()
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouping and summing the data
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Defining a custom color palette
custom_palette = sns.color_palette("Set2", n_colors=len(sales_state))

# Creating the bar plot with the custom palette
sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(data=sales_state, x='State', y='Amount', palette=custom_palette)

# Adding labels and title
plt.title('Total Amount by Top 10 States')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.tight_layout()       # Adjust layout to prevent clipping
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Define a custom color palette
custom_palette = sns.color_palette("Set2", n_colors=df['Occupation'].nunique())

# Create the countplot with the custom palette
ax = sns.countplot(data=df, x='Occupation', palette=custom_palette)

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouping and summing the data
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Defining a custom color palette
custom_palette = sns.color_palette("Set2", n_colors=len(sales_state))

# Creating the bar plot with the custom palette
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=sales_state, x='Occupation', y='Amount', palette=custom_palette)

# Adding labels on top of the bars
ax = plt.gca()
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.title('Total Amount by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.tight_layout()       # Adjust layout to prevent clipping
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*
